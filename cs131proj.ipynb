```python
import pandas as pd
path = "/content/drive/MyDrive/MoviesOnStreamingPlatforms.csv"


# Read the CSV file
df = pd.read_csv(path)

# Remove any row that contains NaN values
df_cleaned = df.dropna()

# Display the cleaned DataFrame
df_cleaned
```

```python
# @title Year

from matplotlib import pyplot as plt
df['Year'].plot(kind='hist', bins=20, title='Year')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Age

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Age').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
